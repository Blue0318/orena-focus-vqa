```python
!pip install orena-focus -q

from focus import FocusDataset, DatasetSplit, Track
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Load FRAME train split
ds = FocusDataset('heico', DatasetSplit.TRAIN, Track.FRAME)
print(f'Total samples: {len(ds)}')
```

```python
# Inspect one sample
request, reference = ds[0]
print('Question:', request.question)
print('Answer:', reference.answer)
print('Format:', reference.format.type if hasattr(reference, 'format') else 'unknown')
print('Request fields:', vars(request).keys())
```

```python
# Distribution of answer formats
formats = []
for req, ref in ds:
    try:
        formats.append(str(ref.format.type))
    except:
        formats.append('unknown')

counter = Counter(formats)
print(counter)

plt.bar(counter.keys(), counter.values())
plt.title('Answer format distribution')
plt.xticks(rotation=30)
plt.show()
```

```python
# Show a sample image
req, ref = ds[0]
if hasattr(req, 'image') and req.image:
    plt.imshow(req.image)
    plt.title(req.question)
    plt.axis('off')
    plt.show()
```
