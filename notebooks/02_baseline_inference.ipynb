```python
import sys
sys.path.append('/kaggle/working/orena-focus-vqa/src')  # adjust path if needed

import yaml
from focus_vqa.data.loader import load_dataset, get_requests_and_references
from focus_vqa.models.qwen_vl import QwenVLModel
from focus_vqa.evaluation.metrics import run_evaluation, show_failures
```

```python
with open('/kaggle/working/orena-focus-vqa/configs/frame.yaml') as f:
    cfg = yaml.safe_load(f)

# Debug run: only 10 samples
cfg['split'] = 'train'
```

```python
ds = load_dataset(cfg)
requests, references = get_requests_and_references(ds)

# Start small
N = 10
requests, references = requests[:N], references[:N]
print(f'Running on {N} samples')
```

```python
model = QwenVLModel(cfg)
model.load()
```

```python
answers = model.predict_batch(requests)

for req, ans, ref in zip(requests, answers, references):
    print(f'Q: {req.question}')
    print(f'Pred: {ans}')
    print(f'Ref:  {ref.answer}')
    print('---')
```

```python
results_df, summary_df = run_evaluation(requests, references, answers, output_dir='results/frame')
show_failures(results_df)
```
